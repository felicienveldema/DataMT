```python
import numpy as np
import pandas as pd
```

```python
# A is data available to use for train set and validation set
# B is data used for test set
```

```python
col_selection = ['srch_id', 'prop_id', \
                 'prop_starrating', 'price_usd', 'prop_review_score', \
                 'prop_location_score1', 'prop_location_score2', 'prop_log_historical_price', \
                 'promotion_flag', 'srch_query_affinity_score', 'cr1', 'cr2', 'random_bool', \
                 'click_bool', 'booking_bool']
```

```python
df_A = pd.read_csv("train1.csv", sep=',')
df_B = pd.read_csv("test1.csv", sep=',')
```

```python
df_A = df_A[col_selection]
df_B = df_B[col_selection]
```
